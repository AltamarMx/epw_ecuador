```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
f = "../data/001_raw/vivienda02/Habitacion principal - sensor denisse - de 30 de noviembre al 6 de enero.xlsx"
recamara1 = pd.read_excel(f,usecols=[1,2,3],names=['tiempo','Ti','hr'],index_col=0,parse_dates=True)
recamara1 = recamara1.add_suffix('_R1')
recamara1

f = "../data/001_raw/vivienda02/Habitacion secundaria - 20809036 - otra fecha - de 12 de enero al 25 de febrero.xlsx"
recamara2 = pd.read_excel(f,usecols=[1,2,3],names=['tiempo','Ti','hr'],index_col=0,parse_dates=True)
recamara2 = recamara2.add_suffix('_R2')
recamara2

f = "../data/001_raw/vivienda02/sala - sensor 20809036  - de 30 de noviembre al 6 de enero.xlsx"
sala = pd.read_excel(f,usecols=[1,2,3],names=['tiempo','Ti','hr'],index_col=0,parse_dates=True)
sala = sala.add_suffix('_sala')
sala
```

```python
viv2 = pd.concat([recamara1,recamara2,sala],axis=0)
viv2
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,6),sharex=True)


f1 = parse('2024-12-01')
f2 = f1 + pd.Timedelta('30D')

ax[0].plot(viv2.Ti_R1,label='R1')
ax[0].plot(viv2.Ti_R2,label='R2')
ax[0].plot(viv2.Ti_sala,label='Sala')

ax[1].plot(viv2.hr_R1,label='R1')
ax[1].plot(viv2.hr_R2,label='R2')
ax[1].plot(viv2.hr_sala,label='Sala')

for axes in ax:
    axes.legend()
    axes.set_xlim(f1,f2)
```
