```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = "../data/001_raw/vivienda01/comedor - sensor dennis - de 5 de noviembre al 22 de noviembre.csv"
comedor = pd.read_csv(f,index_col=0,parse_dates=True,dayfirst=False,usecols=[1,2,3])
comedor.columns = ['Ti_comedor','hr_comedor']

f = "../data/001_raw/vivienda01/dormitorio - sensor 20809036 - de 5 de noviembre al 22 de noviembre.csv"
dormitorio = pd.read_csv(f,index_col=0,parse_dates=True,dayfirst=False,usecols=[1,2,3])
dormitorio.columns = ['Ti_dormitorio','hr_dormitorio']
dormitorio

f = "../data/001_raw/vivienda01/sala - sensor richard - de 5 de noviembre al 22 de noviembre.csv"
sala = pd.read_csv(f,index_col=0,parse_dates=True,dayfirst=False,usecols=[1,2,3])
sala.columns = ['Ti_sala','hr_sala']
```

```python
viv1 = pd.concat([comedor,dormitorio,sala],axis=0)
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,6),sharex=True)

ax[0].plot(viv1.Ti_comedor,label='Comedor')
ax[0].plot(viv1.Ti_dormitorio,label='Dormitorio')
ax[0].plot(viv1.Ti_sala,label='Sala')

ax[1].plot(viv1.hr_comedor,label='Comedor')
ax[1].plot(viv1.hr_dormitorio,label='Dormitorio')
ax[1].plot(viv1.hr_sala,label='Sala')

for axes in ax:
    axes.legend()
```
