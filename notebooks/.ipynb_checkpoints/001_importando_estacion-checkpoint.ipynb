```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = "../data/001_raw/DatosClimaIkiam.csv"
ikiam = pd.read_csv(f, encoding="ANSI",sep=";",skiprows=1)
ikiam.columns = [
    "time",
    "To_2m",
    "hr_2m",
    "Ig_2m",
    "ws_2m",
    "wd_2m",
    "To_10m",
    "hr_10m",
    "Ig_10m",
    "Id_10m",
    "ws_10m",
    "wd_10m",
    "Tg_0cm",
    "Tg_5cm",
    "Tg_10cm",
    "Tg_30cm",
    "Tg_50cm",
    "Tg60cm",
    "Tg80cm"
]
ikiam.time = pd.to_datetime(ikiam.time,format="%d/%m/%Y %H:%M",errors="coerce")
ikiam.set_index("time",inplace=True)
ikiam.index
```

```python
ikiam.plot(subplots=True,figsize=(12,35));
```
